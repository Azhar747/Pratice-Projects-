## Adding Age column to diffrenciate Young , Middle and Old

```python
def givingage(row):
    if row['Rings']>=1 and row['Rings']<=8:
        return 1
    elif row['Rings'] >=9 and row['Rings']<=10:
        return 2
    elif row['Rings']>=11 and row['Rings']<=29:
         return 3
    return 0
```

```python
data['Age']=data.apply (lambda row: givingage(row),axis=1)
```

## Traing Test Split

```python
X=data.drop(['Rings',"Age"],axis=1)
```

```python
y=data[['Age']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logReg=LogisticRegression()
```

```python
logReg.fit(X_train,y_train)
```

```python
y_pred=logReg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
import math
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import median_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import classification_report
acc_score=accuracy_score(y_pred,y_test)
print(acc_score*100)
```

```python
mae=mean_absolute_error(y_pred,y_test);
mae
```

```python
rmse = math.sqrt(mean_squared_error(y_pred,y_test))
rmse
```

```python
report=classification_report(y_pred,y_test)
```

```python
print(report)
```

## Hyperparmater Tuning Logistic Regression

```python
# example of grid searching key hyperparametres for logistic regression
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV
solvers = ['newton-cg', 'lbfgs', 'liblinear']
penalty = ['l2']
c_values = [100, 10, 1.0, 0.1, 0.01]
# define grid search
grid = dict(solver=solvers,penalty=penalty,C=c_values)
grid_search = GridSearchCV(estimator=logReg, param_grid=grid, n_jobs=-1,scoring='accuracy',error_score=0)
grid_result = grid_search.fit(X_train,y_train)
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
